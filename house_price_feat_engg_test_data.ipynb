```python
#find all missing values in categorical features

feature_nan = [feature for feature in housetest.columns if housetest[feature].isnull().sum() > 1 and housetest[feature].dtype == 'O']

feature_nan
```

```python
for feature in feature_nan:
    print('{} has {} % missing values'.format(feature, np.round(housetest[feature].isnull().mean(),4)))
```

```python
def replace_miss_cat(x,y):
    copy2 = housetest.copy()
    copy2[y] = copy2[y].fillna('Missing')
    return copy2

housetest = replace_miss_cat(housetest,feature_nan)

housetest[feature_nan].isnull().sum()
```

```python
#find all missing values in numerical features
# not must to use housetest.columns

numerical_nan = [feature for feature in housetest if housetest[feature].isnull().sum() > 1 and housetest[feature].dtype != 'O']

for feature in numerical_nan:
    print('{} has {}% missing values'.format(feature, np.round(housetest[feature].isnull().mean(),4)))
```

```python
for feature in numerical_nan:
    copy2 = housetest.copy()
    median_values = housetest[feature].median()
    #creating new feature
    housetest[feature + 'nan'] = np.where(housetest[feature].isnull(),1,0)
    housetest[feature].fillna(median_values, inplace = True)
```

### Handling Rare categorical features

```python
categorical_feature = [feature for feature in housetest.columns if housetest[feature].dtype == 'O']
categorical_feature
```

```python
for feature in categorical_feature:
    temp = housetest.groupby(feature)['MSZoning'].count()/len(housetest)
    temp_df = temp[temp > 0.01].index
    housetest[feature] = np.where(housetest[feature].isin(temp_df),housetest[feature],'Rare_var')
```

```python
for feature in categorical_feature:
    labels_ordered = housetest.groupby(feature)['MSSubClass'].mean().sort_values().index
    labels_ordered = {y:x for x,y in enumerate(labels_ordered,0)}
    housetest[feature] = housetest[feature].map(labels_ordered)
```
